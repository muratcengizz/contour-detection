```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Görüntü işlemede kontur algılama, bir nesnenin veya nesne sınırlarının belirlenmesi için kullanılan bir tekniktir. Kontur, bir nesnenin dış hatlarını temsil eden bir eğri veya sınırlayıcı bir çizgidir.

```python
img = cv2.imread(filename="images/contour.jpg", flags=0)
plt.imshow(X=img, cmap="gray");
```

```python
contours, hierarchy = cv2.findContours(image=img, 
                                       mode=cv2.RETR_CCOMP, 
                                       method=cv2.CHAIN_APPROX_SIMPLE)
```

```python
contours[0:2]
```

```python
hierarchy
```

```python
external_contour = np.zeros(shape=img.shape)
internal_contour = np.zeros(shape=img.shape)
```

```python
for element in range(len(contours)):
    # external
    if hierarchy[0][element][3] == -1:
        cv2.drawContours(image=external_contour, contours=contours, contourIdx=element, color=50, thickness=-1)
    else: # internal
        cv2.drawContours(image=internal_contour, contours=contours, contourIdx=element, color=50, thickness=-1)

plt.imshow(X=external_contour, cmap="gray");
```

```python
plt.imshow(X=internal_contour, cmap="gray");
```

## cv2.findContours(image, mode, method, offset)
### Bu fonksiyon, bir görüntüdeki konturları bulmak için kullanılır.
### image: Konturları bulmak için giriş görüntüsü.
### mode: Kontur tespit modu. Bu parametre, kontur bulma yöntemini belirler. Örneğin, "cv2.RETR_EXTERNAL" yalnızca dış konturları bulmak için kullanılırken, "cv2.RETR_TREE" tüm konturları hiyerarşik olarak bulur. Diğer kullanılabilir modlar arasında "cv2.RETR_LIST", "cv2.RETR_CCOMP" ve "cv2.RETR_FLOODFILL" bulunur. Varsayılan değer "cv2.RETR_EXTERNAL"dır.
### method: Kontur yaklaşım yöntemi. Bu parametre, konturun hangi yaklaşım yöntemiyle belirleneceğini belirler. Örneğin, "cv2.CHAIN_APPROX_NONE" tüm kontur noktalarını saklarken, "cv2.CHAIN_APPROX_SIMPLE" noktaları sadece konturunun temsil noktalarıyla temsil eder. Diğer kullanılabilir yöntemler arasında "cv2.CHAIN_APPROX_TC89_L1", "cv2.CHAIN_APPROX_TC89_KCOS" bulunur. Varsayılan değer "cv2.CHAIN_APPROX_SIMPLE"dır.
### offset: İsteğe bağlı bir parametredir. Kontur noktalarının her birini ofsetlemek için kullanılır.

## Bu fonksiyonun dönüş değerleri şunlardır:
### contours: Bulunan konturların listesi. Her kontur, bir dizi noktayı temsil eder.
### hierarchy: Konturların hiyerarşik yapısını içeren bir dizi. Bu dizi, konturların ilişkilerini gösterir.

## cv2.drawContours(image, contours, contourIdx, color, thickness, lineType, hierarchy)
### Bu fonksiyon, bir görüntü üzerine konturları çizmek için kullanılır. 
### image: Konturların çizileceği hedef görüntü.
### contours: Çizilecek konturların listesi. "cv2.findContours()" fonksiyonundan dönen kontur listesi kullanılır.
### contourIdx: Çizilecek konturun indeksi. Eğer negatif bir değer verilirse, tüm konturlar çizilir. Pozitif bir değer verilirse, sadece belirli bir kontur çizilir.
### color: Konturun rengi. Renk değeri BGR formatında verilir.
### thickness: Konturun kalınlığı. Negatif bir değer verilirse, kontur içini doldurur.
### lineType: Kontur çizgisi tipi. Örneğin, "cv2.LINE_8" veya "cv2.LINE_AA" gibi değerler kullanılabilir.
### hierarchy: Konturların hiyerarşik yapısını içeren bir dizi. Bu parametre isteğe bağlıdır.

```python
img = cv2.imread(filename="images/contour.jpg", flags=0)
plt.imshow(X=img, cmap="gray");
```

```python
contours, hierarchy = cv2.findContours(image=img, 
                                       mode=cv2.RETR_EXTERNAL, 
                                       method=cv2.CHAIN_APPROX_NONE)
```

```python
for element in range(len(contours)):
    cv2.drawContours(image=img, 
                     contours=contours, 
                     contourIdx=element, 
                     color=(120, 100, 100), 
                     thickness=3)
    
plt.imshow(X=img, cmap="gray");
```
